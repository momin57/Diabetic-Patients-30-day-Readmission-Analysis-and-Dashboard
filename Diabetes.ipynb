## Exploratory Data Analysis

### Categorical Analysis

```python
categorical_cols = [
    'race', 'gender', 'age', 'age_group', 'medical_specialty',
    'diag_1', 'diag_2', 'diag_3', 'max_glu_serum', 'A1Cresult',
    'metformin', 'repaglinide', 'glimepiride', 'glipizide', 'glyburide',
    'tolbutamide', 'pioglitazone', 'rosiglitazone',
    'insulin', 'glyburide-metformin', 'glipizide-metformin',
    'glimepiride-pioglitazone', 'metformin-rosiglitazone', 'metformin-pioglitazone',
    'change', 'diabetesMed', 'readmitted', 'rare_drugs_used'
]
```

```python
df['readmitted_flag'] = df['readmitted'].apply(lambda x: 1 if x == '<30' else 0)
```

```python
from scipy.stats import chi2_contingency

chi_square_results = []

for col in categorical_cols:
    try:
        contingency = pd.crosstab(df[col], df['readmitted_flag'])
        chi2, p, _, _ = chi2_contingency(contingency)
        chi_square_results.append((col, p))
    except:
        chi_square_results.append((col, 'ERROR'))  # in case of any issues like all 1 value
```

```python
chi_df = pd.DataFrame(chi_square_results, columns=['Feature', 'P-Value'])
chi_df = chi_df[chi_df['P-Value'] != 'ERROR']
chi_df['P-Value'] = chi_df['P-Value'].astype(float)
chi_df = chi_df.sort_values(by='P-Value')

print(chi_df)
```

### âœ… Key Findings of Categorical Analysis:

1. **Strong Associations (p < 0.001):**

   * **Diagnosis Codes (`diag_1`, `diag_2`, `diag_3`)**
     â†’ Diagnosis types have a very strong statistical relationship with readmission rates. Certain diagnoses significantly increase the likelihood of early readmission.

   * **Insulin Usage**
     â†’ The status of insulin (e.g., up/down/steady) during the visit is highly associated with readmission. This suggests insulin adjustments may be a clinical signal for risk.

   * **Medical Specialty**
     â†’ Patients treated under specific specialties (e.g., internal medicine, emergency medicine) show significantly different readmission patterns.

   * **Age and Age Group**
     â†’ Older patients (especially those grouped as 'Senior' or 'Elderly') tend to have higher 30-day readmission rates.

   * **Diabetic Medication (`diabetesMed`)**
     â†’ Whether or not diabetic medications were prescribed is a strong predictor of readmission.

   * **Metformin & Medication Change**
     â†’ Use of metformin, and whether there was a change in medication (`change` column), both show strong associations.

---

2. **Moderate Associations (0.001 < p < 0.05):**

   * **Race**
     â†’ Small but statistically significant differences in readmission rates exist across racial groups.
   * **Some individual medications**: `repaglinide`, `glipizide`
     â†’ These specific medications had weak but statistically notable associations with readmission.

---

3. **No Significant Association (p > 0.05):**

   * **Gender**
     â†’ No meaningful difference in readmission rates between male and female patients.
   * **Rare Drugs Used**
     â†’ Grouped rare drugs do not significantly correlate with readmission when considered alone.
   * **Combination Medications**
     â†’ Most combination diabetes medications (e.g., `metformin-pioglitazone`, `glipizide-metformin`) showed no statistical relationship with readmission.
   * **A1Cresult**
     â†’ Surprisingly, A1C test results did not show a significant difference in predicting readmission.

```python
sns.set(style="whitegrid")

def plot_bar(column, title=None, figsize=(12, 6)):
    plt.figure(figsize=figsize)
    ax = sns.barplot(
        x=column,
        y='readmitted_flag',
        data=df,
        estimator=lambda x: sum(x) / len(x),  # Mean = % readmitted
        ci=None
    )
    ax.set_title(title or f'Readmission Rate by {column}')
    ax.set_ylabel('Proportion Readmitted (<30)')

    # Make labels vertical and smaller
    plt.xticks(rotation=90, fontsize=8)
    
    plt.tight_layout()
    plt.show()
```

### âœ… Key Findings of Numerical Analysis:

1. **number\_inpatient**

   * Strongest predictor of readmission across all tests.
   * Significantly higher in the readmitted group (mean: 1.09 vs 0.36).
   * Highest correlation with readmission (`r = 0.165`).

   > Patients with more prior inpatient visits were much more likely to be readmitted within 30 days.

2. **number\_diagnoses**

   * Moderate indicator with a slight increase in readmitted patients.
   * Statistically significant with a small effect size (`r = 0.045`).

   > More diagnoses suggest higher comorbidity burden, mildly associated with readmission.

3. **number\_emergency**

   * Statistically significant but sparse data with many zeros.
   * Correlation is moderate (`r = 0.061`); outliers contribute to the signal.

   > Frequent emergency visits may indicate instability and higher readmission risk.

4. **time\_in\_hospital**

   * Mean hospital stay was slightly longer for readmitted patients.
   * Statistically significant (`p < 0.05`), but weak correlation (`r = 0.044`).

   > Small increase in length of stay does not strongly predict readmission.

5. **num\_medications**

   * Readmitted patients were on slightly more medications.
   * Weak correlation (`r = 0.038`), but statistically significant.

   > Polypharmacy may contribute slightly to readmission, though effect is small.

6. **num\_lab\_procedures**

   * Very similar between groups with minimal difference.
   * Weak correlation (`r = 0.023`) despite significant p-value due to large sample size.

   > Not clinically meaningful as a predictor of readmission.

7. **number\_outpatient**

   * Mostly zero values; statistically significant but practically uninformative.
   * Correlation is very weak (`r = 0.019`).

   > Sparse data limits the usefulness of this feature in modeling.

8. **num\_procedures**

   * No difference observed, not statistically significant (`p = 0.505`).
   * Slight negative correlation (`r = -0.012`).

   > Does not contribute to distinguishing readmitted patients.
